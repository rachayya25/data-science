```python
import pandas as pd
```

```python
test = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
# 1.0 Number of male and female passengers
male_pass = test[test['Sex']== 'male']
female_pass = test[test['Sex'] == 'female']
len(male_pass)
```

```python
len(female_pass)
```

```python
# 2.0 number of people survived
survived = test[test['Survived']==1]
len(survived)
```

```python
# 3.0 Number of casulaties
casualties = test[test['Survived']==0]
len(casualties)
```

```python
# Name of eldest person
eldest_name = test[test['Age'] == max(test['Age'])] ['Name']
eldest_name
```

```python
# 5.0 number of passengers in pclass 1, 2, 3
pclass1 = test[test['Pclass'] == 1]
pclass2 = test[test['Pclass'] == 2]
pclass3 = test[test['Pclass'] == 3]
[len(pclass1), len(pclass2),len(pclass3)]
```

```python
# 6.0 person name starts with 's'
name_s = test['Name']
count = 0
for i in name_s:
    if i[0]=='S' or i[0]=='s'  :
        count += 1
print(count)
```

```python
# 8.0 number of person's below age 25
age_25 = test[test['Age'] < 25]
len(age_25)
```

```python
# 9.0 persons did below age 40
died_40 = test[test['Survived']==0] ['Age']
count = 0
for i in died_40:
    if i < 40:
        count += 1
print(count)
```
